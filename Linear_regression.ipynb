```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model 
from sklearn.metrics import mean_squared_error
```

```python
#load the preexisting data into it
diabetes = datasets.load_diabetes()
```

```python
print(diabetes.keys())
#['data', 'target', 'frame', 'DESCR', 'feature_names', 'data_filename', 'target_filename', 'data_moduleh']
print(diabetes.DESCR)
```

```python
diabetes_X = diabetes.data[:, np.newaxis, 2]
print(diabetes_X)
```

```python
# dividing the data into train and test data
diabetes_X_train = diabetes_X[:-30] #features
diabetes_X_test = diabetes_X[-20:]
```

```python
diabetes_y_train = diabetes.target[:-30] #labels
diabetes_y_test = diabetes.target[-20:]
```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_X_train,diabetes_y_train)
diabetes_y_predict = model.predict(diabetes_X_test)
```

```python
print("Mean square error is :",mean_squared_error(diabetes_y_test,diabetes_y_predict) )
```

```python
#coefficient of the values are given out
print("Weights:" , model.coef_) 
print("Intercept: ", model.intercept_)
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.show()
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.plot(diabetes_X_test,diabetes_y_predict)
plt.show()
```

```python
#Now consider all the features 
diabetes_X = diabetes.data
diabetes_X_train = diabetes_X[:-30] #features
diabetes_X_test = diabetes_X[-20:]
diabetes_y_train = diabetes.target[:-30] #labels
diabetes_y_test = diabetes.target[-20:]
model = linear_model.LinearRegression()
model.fit(diabetes_X_train,diabetes_y_train)
diabetes_y_predict = model.predict(diabetes_X_test)
print("Mean square error is :",mean_squared_error(diabetes_y_test,diabetes_y_predict) )
print("Weights:" , model.coef_) 
print("Intercept: ", model.intercept_)
# Intercept is considered as W0 
```
